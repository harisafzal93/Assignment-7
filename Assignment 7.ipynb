```python
class employee:
    def __init__(self, name, number, address, salary, post, year_of_joining, education, years_of_experience):
        self.name = name
        self.number = number
        self.address = address
        self.salary = salary
        self.post = post
        self.year_of_joining = year_of_joining
        self.education = education
        self.years_of_experience = years_of_experience

    def describe_employee(self):
        print(f"Employee Name: {self.name}")
        print(f"Employee Number: {self.number}")
        print(f"Address: {self.address}")
        print(f"Salary: {self.salary}")
        print(f"Post: {self.post}")
        print(f"Year of Joining: {self.year_of_joining}")
        print(f"Education: {self.education}")
        print(f"Years of Experience: {self.years_of_experience}")

    def update_years_of_experience(self, additional_years):
        self.years_of_experience += additional_years
        print(f"Years of experience updated to: {self.years_of_experience} years")

    def increase_salary(self, increment):
        self.salary += increment
        print(f"Salary increased by ${increment}. New salary is ${self.salary}.")

employee101 = employee(
    name="Naveed Ahmed",
    number=792976,
    address="DHA phase II, Karachi, Pakistan",
    salary=75000,
    post="Softwear Engineer",
    year_of_joining=2018,
    education="Bachelor's Degree in Computer Science",
    years_of_experience=5
)

print(employee101.name)
print(employee101.number)
print(employee101.address)
print(employee101.salary)
```

```python
employee101.describe_employee()
employee101.increase_salary(5000)
employee101.update_years_of_experience(2)
```

```python
employees = [
    employee(f"ABC {i}", f"112233{i}", f"Address {i}", 5000 + i * 5000, "Developer", 2015 + i, "Bachelor's", 3 + i)
    for i in range(10)
]

for emp in employees:
    emp.describe_employee()
    emp.increase_salary(3000)
    emp.update_years_of_experience(1)
```

```python
def describe_employee(employee):
    print(f"Employee Name: {employee['name']}")
    print(f"Employee Number: {employee['number']}")
    print(f"Address: {employee['address']}")
    print(f"Salary: ${employee['salary']}")
    print(f"Post: {employee['post']}")
    print(f"Year of Joining: {employee['year_of_joining']}")
    print(f"Education: {employee['education']}")
    print(f"Years of Experience: {employee['years_of_experience']} years")
    
def update_years_of_experience(employee, additional_years):
    employee['years_of_experience'] += additional_years
    print(f"Years of experience updated to: {employee['years_of_experience']} years")

def increase_salary(employee, increment):
    employee['salary'] += increment
    print(f"Salary increased by ${increment}. New salary is ${employee['salary']}.")

employee101 = {
    "name": "Ali Hassan",
    "number": "210485",
    "address": "PECHS BLOCK 02, Karachi, Pakistan",
    "salary": 50000,
    "post": "Banker",
    "year_of_joining": 2022,
    "education": "Bachelor's Degree in Commerce",
    "years_of_experience": 5
}

print(employee101['name'])
print(employee101['number'])
print(employee101['address'])
print(employee101['salary'])
```

```python
describe_employee(employee101)
increase_salary(employee101, 5000)
update_years_of_experience(employee101, 2)
```
